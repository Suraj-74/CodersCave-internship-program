## Data preprocessing

```python

## check duplicates
df.duplicated().sum()
```

```python
# drop the duplicates values
df.drop_duplicates(inplace=True)
df
```

```python
# check, duplicates values are still there or not
df.duplicated().any()
```

```python
df.shape
```

```python
# extract numerical columns
num_col=df.select_dtypes(exclude="object").columns
num_col
```

```python
# count of missing values with 97
df[df==97].count()
```

```python
# count of missing values with 98
df[df==98].count()
```

```python
# count of missing values with 99
df[df==99].count()
```

```python
df['SEX'].replace({1:'Female', 2: 'Male'}, inplace = True)
df['SEX'].value_counts()
```

```python
# conevrting numerical column(patient_type) into categorical column 
df['PATIENT_TYPE'].replace({1:'Returned Home', 2:'Hospitalization'}, inplace = True)
df['PATIENT_TYPE'].value_counts()
```

```python
# fill missing value with alive and value which is not missing with died
df.DATE_DIED[df['DATE_DIED'] != '9999-99-99'] = 'Died'
df.DATE_DIED[df['DATE_DIED'] == '9999-99-99'] = 'Alive'
df.DATE_DIED.value_counts()
```

```python
# Adding a column
df['DEATH'] = df['DATE_DIED']
```

```python
# Dropping a DATE_DIED Column as it is same as DEATH Column
df.drop(['DATE_DIED'],axis=1,inplace=True) 
```

```python
# Checking Data
df.head()
```

```python
# now check the categorical column again
cat_col=df.select_dtypes(include="object").columns
cat_col
```

```python
## check the null values
df.isnull().sum()
```

```python
# replace the missing values(97) with null values
df.replace(97,np.nan,inplace=True)
df[df==97].count()  # check 
```

Now, there are no missing values with 97

```python
# replace the missing value(98) with null values
df.replace(98, np.nan, inplace=True)
df[df==98].count() # check
```

```python
# replace the missing values(99) with null values
df.replace(99, np.nan, inplace=True)
df[df==99].count()    # check
```

## Feature Selection

```python
# Drop unnecessary column INTUBED which is not in use
 
df.drop("INTUBED", axis=1, inplace=True)
df.head()
```

```python
df.drop(["COPD","INMSUPR","OBESITY","TOBACCO"], axis=1, inplace=True)
df.head()
```

```python
# now check null values
df.isnull().sum()
```

```python
# create a list of columns to drop null values from
columns_to_drop = ['PNEUMONIA', 'AGE', 'PREGNANT', 'DIABETES', 'ASTHMA','HIPERTENSION','OTHER_DISEASE','CARDIOVASCULAR','RENAL_CHRONIC','ICU']

# drop rows with null values in the specified columns
df.dropna(subset=columns_to_drop)
```

```python
# check the null values
df.isnull().sum()
```

Univariate Analysis

```python
# plot boxplot for whole data

df.boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
# plot a histplot 

plt.figure(figsize=(30,25))
df.hist()
plt.show()
```

```python
# counts of survival status

a1=df["DEATH"].value_counts()
a1
```

```python
# percentage distribution of survival status

plt.figure(figsize=(8,6))
sns.set_palette("Set2")
plt.pie(a1.values,labels=a1.index,autopct="%.2f%%")
plt.title("DEATH count percentage distribution")
plt.legend()
plt.show()
```

```python
# kde plot for age column

sns.kdeplot(x=df["AGE"],color="blue")
plt.title("KDE plot for age column")
plt.show()
```

```python
# Plot a bar chart of the patient type variable

df["PATIENT_TYPE"].value_counts().plot(kind="bar",color="pink")
plt.xlabel("Patient Type")
plt.ylabel("Frequency")
plt.title("Distribution of Patient Type")
plt.show()
```

```python
# plot a hist graph for distribution of age
plt.hist(df["AGE"], bins=20, color="cyan")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Age")
plt.show()
```

```python
sns.countplot(x= df['ICU'])
plt.title('Distribution of Patients Admitted to ICU',fontsize=15)
plt.xlabel('Admitted to ICU')
plt.ylabel(' No. of Patients')
plt.show()
```

BIVARIATE ANALYSIS

```python
# number of counts of pregnant on the basis of patient type
a3=df.groupby("PATIENT_TYPE")["PREGNANT"].value_counts()
a3
```

```python
# plot a bar graph 
plt.figure(figsize=(9,8))
cmap="tab10"
a3.plot(kind="bar",cmap=cmap)
plt.title("counts of pregnant on the basis of patient type")
plt.show()
```

```python
# plot boxplot graph for  death with respect to age
plt.figure(figsize=(8,6))
sns.boxplot(data=df,x="DEATH", y="AGE", palette=sns.color_palette(['#90EE90', '#FF0000']))
plt.title("Death with respect to age")
plt.ylabel("Age", fontsize=14)
plt.xticks([0,1],["Died","Alive"], fontsize=13)
plt.show()
```

```python
# plot a countplot of death with respect to gender
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="SEX", hue="DEATH")
plt.title("death with respect to gender",fontsize=20,color='y')
plt.xlabel("SEX", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks([0,1],["Female","Male"], fontsize=13)
plt.show()
```

```python
# Plot a scatter plot of age against pneumonia
sns.scatterplot(x="AGE", y="PNEUMONIA", data=df, color="green")
plt.xlabel("Age")
plt.ylabel("Pneumonia")
plt.title("Age vs. Pneumonia")
plt.show()
```

```python
# correlation data
corr=df.corr()
corr
```

```python
# draw a heatmap for correlation
plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=True, cmap="plasma")
plt.show()
```
