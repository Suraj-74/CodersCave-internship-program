```python
# fill the null value with appropriate value using aggregate function such as mean, median and mode

median_ratings=data["Rating"].median()
median_ratings
```

```python
# remove the null values in rating column
data["Rating"].fillna(median_ratings,inplace=True)
```

```python
# now, there are no null values in rating column
data.isnull().sum()
```

```python
# mode of categorical columns

print(data["Type"].mode())
print(data["Current Ver"].mode())
print(data["Android Ver"].mode())
```

```python
data["Type"].fillna(str(data["Type"].mode().values[0]),inplace=True)
data["Current Ver"].fillna(str(data["Current Ver"].mode().values[0]),inplace=True)
data["Android Ver"].fillna(str(data["Android Ver"].mode().values[0]),inplace=True)
```

```python
data.isnull().sum()
```

```python
# let's convert price, review and rating into numerical value

data["Price"]=data["Price"].apply(lambda x : str(x).replace("$"," ") if "$" in str(x) else str(x))
data["Price"]=data["Price"].apply(lambda x : float(x))
data["Reviews"]=pd.to_numeric(data["Reviews"],errors="coerce")
```

```python
# let's convert installs into numerical value

data["Installs"]=data["Installs"].apply(lambda x : str(x).replace("+"," ") if "+" in str(x) else str(x))
data['Installs'] = data['Installs'].str.replace(' ', '').str.replace(',', '')
data["Installs"]=data["Installs"].apply(lambda x : float(x))
```

```python
# check how many duplicates are

data.duplicated().sum()
```

```python
# drop duplicates value

data.drop_duplicates(inplace=True)
data
```
